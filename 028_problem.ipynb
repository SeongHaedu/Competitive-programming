・神がちょうど k 枚重なっている部分の面積



【制約】

・1 ≦ N ≦ 10^5

・0 ≦ lx[i] < rx[i] ≦ 1000

・0 ≦ ly[i] < ry[i] ≦ 1000

・入力はすべて整数




### 入力形式

N

lx[1] ly[1] rx[1] ry[1]

lx[2] ly[2] rx[2] ry[2]

 :
 
lx[N] ly[N] rx[N] ry[N]
